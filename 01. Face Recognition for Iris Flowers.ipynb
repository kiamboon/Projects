<a id="2.7"><h3>2.7 Central Tendency Measures</h3></a>
- We added the general descriptive statistics like median and variance to the describe() method for the Central Tendency Measure as describe method does not included these two statistical 
- The Median for the middle number of the DataFrame.<br>
- The variance tells us how the values of a variable distribute around the mean. The higher the variance, the farther we can expect some values to appear from the mean.

```python
def describe(df, stats):
    i = df.describe()
    return i.append(df.reindex(i.columns, axis = 1).agg(stats))   # Note: reindex_axis() method is deprecated.

describe(df, ["median", "var"])
```

### The mean and median are the first measures to calculate for numeric variable when starting EDA. The median provides the central position in the series of values. We noticed here that the means here are not centered which is no variable is zero mean and that the median of petal length is quite different from the mean.

<a id="2.8"><h3>2.8 Measuring Range</h3></a>
- In addition, we also check the range, which is the difference between the maximum and minimum value for each quantitative variable, and it is very informative about the difference in scale among variables.

```python
df.max(numeric_only = True) - df.min(numeric_only = True)
```

### ⚠ We noticed that the petal length's standard deviation and range are high with respect to the measures of centrality that may point to a possible problem with extremely unusual values affecting the calculations or an unexpected distribution of values around the mean.

<a id="2.9"><h3>2.9 Testing Measure of Normality</h3></a>
* We noticed petal length is high in standard deviation and range. Here we will check on its skewness.

```python
s = skew(df["petal_length (cm)"])
zscore, pvalue = skewtest(df["petal_length (cm)"])

print("Skewness", "{:.4f}".format(s), "\n", "z-score", "{:.4f}".format(zscore), "\n", "p-value", "{:.4f}".format(pvalue))
```

### ⚠ The result tells us that the petal length is slighyly skewed to the left.

<a id="3.3"><h3>3.3 Univariate Analysis</h3></a>

* Univariate as the name itself means litterally one or single variable, we will use petal length this variable where we found it from above the analysis as univariate analysis.

<a id="3.3.1"><h4>3.3.1 Cumulative Distribution Function (CDF)</h4></a>

* Now we have the most distinguishable variable which is petal length, we can use this variable for cumulative Distribution Function which is one-variable analysis. Next we can use the petal length for 3 type of flowers on CDF.

```python
setosa = df.loc[lambda x: x["class"] == "Setosa"]
versicolor = df.iloc[(df["class"] == "Versicolor").values]
virginica = df.loc[df["class"].isin(["Virginica"])]


plt.figure(figsize = (15, 5))
# Setosa
counts, bins = np.histogram(setosa["petal_length (cm)"], bins = 10, density = True)
pdf = counts / sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bins[1:], pdf, label = "Setosa PDF")
plt.plot(bins[1:], cdf, label = "Setosa CDF")


# Versicolor
counts, bins = np.histogram(versicolor["petal_length (cm)"], bins = 10, density = True)
pdf = counts / sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bins[1:], pdf, label = "Versicolor PDF")
plt.plot(bins[1:], cdf, label = "Versicolor CDF")


# Virginica
counts, bins = np.histogram(virginica["petal_length (cm)"], bins = 10, density = True)
pdf = counts / sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bins[1:], pdf, label = "Virginica PDF")
plt.plot(bins[1:], cdf, label = "Virginica CDF")
plt.legend()
plt.show()
```

### ⚠ Petal length of CDF plot

<a id="5.1.1"><h4>5.1.1 Applying Multinomial Naïve Bayes Model</h4></a>
* The multinomial Naive Bayes Model is suitable for classification with discrete features for exmaple, word counts for text classification, but we will see how good it works on this dataset.

```python
nb = MultinomialNB()
nb.fit(X_train, y_train)

y_pred_nb = nb.predict(X_test)

# Accuracy
score1 = accuracy_score(y_pred_nb, y_test)
print("Accuracy:", "{:.2%}".format(score1), "\n")


# Report and confusion matrix
print(classification_report(y_test, y_pred_nb, target_names = ["Setosa", "Versicolor", "Virginica"]))
cm1 = confusion_matrix(y_test, y_pred_nb)
print(cm1, "\n")


ax = plt.subplot()
sns.heatmap(cm1, annot = True, ax = ax, cmap = "tab20c", fmt = "d")  # d = decimal integer
ax.set_xlabel("(Predicted)")
ax.set_ylabel("(Actual)")
ax.set_title("Confusion Matrix - Naive Bayes Model of 4 Features on the Iris Flowers")
ax.xaxis.set_ticklabels(["Setosa", "Versicolor", "Virginica"])
ax.yaxis.set_ticklabels(["Setosa", "Versicolor", "Virginica"])
plt.show()
```

<a id="5.1.2"><h4>5.1.2 Applying Random Forest Classifier</h4></a>

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)


# Accuracy
score2 = accuracy_score(y_pred_rf, y_test)
print("Accuracy:", "{:.2%}".format(score2), "\n")


# Report and confusion matrix
print(classification_report(y_test, y_pred_rf, target_names = ["Setosa", "Versicolor", "Virginica"]))
cm2 = confusion_matrix(y_test, y_pred_rf)
print(cm2, "\n")


ax = plt.subplot()
sns.heatmap(cm2, annot = True, ax = ax, cmap = "tab20c", fmt = "d")  # d = decimal integer
ax.set_xlabel("(Predicted)")
ax.set_ylabel("(Actual)")
ax.set_title("Confusion Matrix - Random Forest Classifier of 4 Features on the Iris Flowers")
ax.xaxis.set_ticklabels(["Setosa", "Versicolor", "Virginica"])
ax.yaxis.set_ticklabels(["Setosa", "Versicolor", "Virginica"])
plt.show()
```

<a id="5.1.5"><h4>5.1.5 Applying K-Nearest Neighbors Model (KNN)</h4></a>
